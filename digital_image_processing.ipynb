# **Draw a polygon on black image**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Create a black image
black_image = np.zeros((512, 512, 3), np.uint8)

# Define the polygon vertices
pts = np.array([[100, 100], [200, 50], [300, 150], [250, 250]], np.int32)
pts = pts.reshape((-1, 1, 2))

# Draw the polygon
cv2.polylines(black_image, [pts], True, (0, 255, 0), 3) # True closes the polygon

# Display the image
cv2_imshow(black_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Draw a polygon on white image

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Create a white image
white_image = np.ones((512, 512, 3), np.uint8)*255

# Define the polygon vertices
pts = np.array([[100, 100], [200, 50], [300, 150], [250, 250]], np.int32)
pts = pts.reshape((-1, 1, 2))

# Draw the polygon
cv2.polylines(white_image, [pts], True, (0, 255, 0), 3) # True closes the polygon

# Display the image
cv2_imshow(white_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# cv2_imshow(white_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
